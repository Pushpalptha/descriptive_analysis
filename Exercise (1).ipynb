```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("C:/Users/user/Downloads/titanic.csv")
df
```

```python
# Relevent columns names are Age and Sex
df_subset = df[['Age', 'Sex']]
df_subset
```

```python
df_subset = df[['Age', 'Sex']].dropna(subset=['Age'])
df_subset
```

```python
# Relevent variables names are male_ages and female_ages
male_ages = df[df['Sex'] == 'male']['Age']
female_ages = df[df['Sex'] == 'female']['Age']

plt.hist([male_ages, female_ages], bins=30, label=['Male', 'Female'], stacked=True, alpha=0.7)

plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# By using Seaborn library
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Age", hue="Sex", bins=30, multiple="stack", kde=False, alpha=0.7)

plt.title("Age Distribution by Gender")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
